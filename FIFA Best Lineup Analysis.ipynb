# FIFA Best Lineup Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('FullData.csv') # import data
data.head(10)
```

```python
data.shape
```

# Cleaning and Filtering Data

```python
data.drop(['National_Kit', 'Club','Club_Kit','Club_Joining','Contract_Expiry'], axis=1, inplace=True) #Deleting unwanted columns 

data.head(10)
```

```python
plt.figure(figsize=(15,32)) # Sets size of figure
sns.countplot(y = data.Nationality,palette='Set2') # draws countplot
```

```python
data.isnull().sum(axis=0) # checking null values
```

```python
data= data.dropna() # Deleting all null value row 
data
```

# Visualization

```python
plt.figure(figsize=(5,10)) # Sets size of figure
sns.countplot(y = data.Nationality,palette='Set2') # draws countplot of players who play for their nation 
```

```python
plt.figure(figsize=(15,5)) # Sets size of figure
sns.countplot(x = data.Age) # draws countplot of national players age
```

# Best Goalkeeper

```python
# weightage

a=1 # Less Imortant
b=2 # Important
c=3 # Very Important
d=4 # Most Important
```

```python
# Parameters of good goalkeeper

data['GK_Stopper'] =( c*data.Ball_Control + c*data.Marking + c*data.Sliding_Tackle + a*data.Standing_Tackle + c*data.Aggression + d*data.Reactions +d*data.Interceptions + d*data.Vision + b*data.Composure + c*data.Speed + b*data.Stamina + d*data.Strength + d*data.Balance + c*data.Agility + d*data.Jumping + d*data.Heading +b*data.GK_Positioning  + d*data.GK_Diving + a*data.GK_Kicking + d*data.GK_Handling + d*data.GK_Reflexes)
data['GK_Clearer'] =( d*data.Ball_Control + d*data.Dribbling + a*data.Marking + a*data.Sliding_Tackle + a*data.Standing_Tackle + c*data.Aggression + c*data.Reactions + b*data.Attacking_Position + b*data.Interceptions + b*data.Vision + b*data.Composure + d*data.Crossing + d*data.Short_Pass + d*data.Long_Pass + d*data.Acceleration + d*data.Speed + c*data.Stamina + c*data.Strength + b*data.Balance + b*data.Agility + a*data.Jumping + b*data.Heading + d*data.Shot_Power + d*data.Long_Shots + a*data.Curve + d*data.Freekick_Accuracy + b*data.GK_Positioning + b*data.GK_Diving + d*data.GK_Kicking + c*data.GK_Handling + d*data.GK_Reflexes )
```

```python
plt.figure(figsize=(10,5))

Sorted_data= data.sort_values('GK_Stopper', ascending=False)
Sort_A=Sorted_data[Sorted_data['Club_Position']=='GK'][:5]

x1=np.array(list(Sort_A['Name']))
y1=np.array(list(Sort_A['GK_Stopper']))

sns.barplot(x1,y1)
plt.title('Best goal Stopper Goalkeeper')
plt.xlabel('Name')
plt.ylabel('Score')
```

```python
plt.figure(figsize=(10,5))

Sorted_data= data.sort_values('GK_Clearer', ascending=False)
Sort_A=Sorted_data[Sorted_data['Club_Position']=='GK'][:5]

x1=np.array(list(Sort_A['Name']))
y1=np.array(list(Sort_A['GK_Clearer']))

sns.barplot(x1,y1)
plt.title('Best Ball Clearer Goalkeeper')
plt.xlabel('Name')
plt.ylabel('Score')
```

```python
# Both graph shows MANUEL NEUER is the best clearer and stopper goalkeeper
pd.set_option('display.max.columns',None)
Sort_A[Sort_A['Name']=='Manuel Neuer']
```
